# Introduction

## __call__()
When you write a function and invoke it, this is what is executed internally by the interpreter

```python
def addition(a,b):
    return (a+b)

c = addition(10,20)  <---> c = addition.__call__(10,20)
```

## Example1

`````python
def addition(a,b):
    return (a+b)
`````

`````python
print(addition(10,20))
print(addition.__call__(10,20))
`````

## Where to use? 
- func() is easier than func.__call__()
- So, it is used in Classes and Objects
> __call__() makes an object behave like a function

## Class And Object Usage

```python
class Hello():
    async def __call__():

hello = Hello()                 # Create an object
hello()                         # Use it like a function
````

`````python
class Greeter:
    def __init__(self, greeting : str = "Hello"):
        self.greeting = greeting
    
    def __call__(self, name : str = "Default User"):
        return f"{self.greeting} {name}!"
`````

`````python
greeter = Greeter()
greeter.greeting
`````

`````python
hi_greeter = Greeter("Hi")
hi_greeter.greeting
`````

`````python
print(hi_greeter())
print(hi_greeter("Anuj"))
`````

    def __init__(self, auto_error = True):

        creds = await super().__call__(request)
        token = creds.credentials
        payload = decode_access_token(token)
